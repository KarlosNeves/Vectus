```python
# 04_comparacao.ipynb ‚Äî Simulado vs Real
import pandas as pd
import matplotlib.pyplot as plt

# Carrega os dados
sim = pd.read_csv('output/simulacoes_com_ict.csv')
real = pd.read_csv('Pasted_Text_1764254826189.txt')

# Verifica colunas-chave
print("‚úÖ Simulado: ", sim.shape, "| Colunas:", list(sim.columns))
print("‚úÖ Real:     ", real.shape, "| Colunas:", list(real.columns))

# Adiciona coluna 'origem'
sim['origem'] = 'Simulado'
real['origem'] = 'Real'

# Junta os dados
df = pd.concat([sim, real], ignore_index=True)

# Estat√≠sticas por origem
stats = df.groupby('origem').agg(
    ict_media=('ict', 'mean'),
    carga_media=('carga_efetiva' if 'carga_efetiva' in df.columns else 'total_tributos/valor_fob_brl', 'mean'),
    eficiente_pct=('classificacao_ict', lambda x: (x == '‚úÖ Eficiente').mean() * 100),
    moderado_pct=('classificacao_ict', lambda x: (x == '‚ö†Ô∏è Moderado').mean() * 100),
    total_ops=('ict', 'size')
).round(2)

print("\nüìä Compara√ß√£o Simulado √ó Real:")
print(stats)

# Gr√°fico: distribui√ß√£o de ICT
plt.figure(figsize=(10, 6))
plt.hist(sim['ict'], bins=30, alpha=0.6, label='Simulado', color='steelblue', edgecolor='white')
plt.hist(real['ict'], bins=30, alpha=0.6, label='Real', color='orange', edgecolor='white')
plt.axvline(5, color='green', linestyle='--', label='Limite Eficiente (ICT=5)')
plt.axvline(15, color='red', linestyle='--', label='Limite Cr√≠tico (ICT=15)')
plt.xlabel('√çndice de Complexidade Tribut√°ria (ICT)')
plt.ylabel('Frequ√™ncia')
plt.title('Distribui√ß√£o de ICT: Simulado √ó Real')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('output/comparacao_ict.png', dpi=200)
plt.show()
```
