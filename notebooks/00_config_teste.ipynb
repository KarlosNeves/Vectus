```python
# 00_config_teste.ipynb
import json
import os

# Caminho do notebook atual
notebook_path = os.path.abspath("__file__") if "__file__" in globals() else os.getcwd()

# Se estiver no Jupyter, __file__ n√£o existe ‚Üí usar caminho da pasta atual
if "__file__" not in globals():
    notebook_dir = os.getcwd()
else:
    notebook_dir = os.path.dirname(notebook_path)

# Caminho para config.json (ele t√° na pasta PAI, n√£o em notebooks/)
config_path = os.path.join(notebook_dir, "..", "config.json")
config_path = os.path.normpath(config_path)  # ajusta barras no Windows/macOS

print("üîç Procurando config.json em:", config_path)
print("‚û°Ô∏è Arquivo existe?", os.path.exists(config_path))

if os.path.exists(config_path):
    with open(config_path, encoding='utf-8') as f:
        cfg = json.load(f)
    print("\n‚úÖ Config carregado com sucesso!")
    print(f"‚Ä¢ Simula√ß√µes: {cfg['simulacoes']}")
    print(f"‚Ä¢ Benchmark ICT: {cfg['benchmark_ict']}")
    print(f"‚Ä¢ Pasta de sa√≠da: '{cfg['output_dir']}'")
else:
    print("\n‚ùå Arquivo n√£o encontrado. Verifique:")
    print("   - Se 'config.json' est√° na pasta raiz (tributec-ai/)")
    print("   - Se voc√™ n√£o est√° em uma subpasta como 'notebooks/' sem ajustar o caminho")
```
