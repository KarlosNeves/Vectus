```python
import pandas as pd

print("üîç Carregando 10.000 empresas (amostra r√°pida)...")

df = pd.read_csv(
    "../data/empresas_rf.txt",
    sep=';',
    encoding='latin1',
    header=None,
    names=[
        'cnpj_basico',
        'razao_social',
        'natureza_juridica',
        'qualificacao_responsavel',
        'capital_social',
        'porte_empresa',
        'ente_federativo'
    ],
    on_bad_lines='skip',      # ‚Üê ignora linhas com erro (comum na RF)
    nrows=10_000,             # ‚Üê s√≥ 10 mil linhas (r√°pido!)
    low_memory=False          # ‚Üê evita warning de dtype
)

# Converte capital_social com seguran√ßa
df['capital_social'] = pd.to_numeric(
    df['capital_social'].str.replace(',', '.'),
    errors='coerce'
).fillna(0.0)

print("‚úÖ Carregado!")
print("Shape:", df.shape)
df.head()
```
